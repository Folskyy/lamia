# Importações

```python
import shap
import sklearn
```

# Processamento

Dataset com informações de casas na California

```python
x, y = shap.datasets.california()
```

```python
lim = int(round(len(x) * 0.20))
```

```python
x_train, y_train = x[:lim], y[:lim]
x_test, y_test = x[lim:], y[lim:]
```

# Modelo

```python
model = sklearn.linear_model.LinearRegression()
```

```python
model.fit(x_train,y_train)
```

# SHAP

```python
explainer = shap.Explainer(model.predict, x_train)
shap_values = explainer(x_test)
```

Gráfico em cascata para visualizar os valores SHAP da primeira amostra do 'x_test'

```python
shap.plots.waterfall(shap_values[0])
```

Gráfico que resume a distribuição dos impactos de todas as features usadas pelo modelo em suas previsões.

```python
shap.summary_plot(shap_values, x_test)
```
